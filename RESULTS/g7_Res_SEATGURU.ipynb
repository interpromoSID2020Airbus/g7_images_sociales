```python
import pandas as pd
```

```python
path_pred = 'pred/'
path_real = 'df_csv/'
```

```python
df_pred = pd.read_csv(path_pred + 'pred_SEATGURU.csv', sep=';')
df_real = pd.read_csv(path_real + 'g7_SEATGURU_annotate.csv', sep=';')
```

```python
df_pred.rename(columns={"img": "name"}, inplace = True)
```

```python
df = pd.merge(df_real, df_pred, on = "name")
```

```python
df
```

```python
ls = []
ls2 =[]
ls3 = []
ls4 = []
ls_model = ['A320','A321','A330','A350','A380','737','747','757','777']
ls_man = ['Airbus', 'Boeing']

for k in range(len(df.loc[:, 'name'])):
    if df.loc[:, 'view_x'][k] != df.loc[:, 'view_y'][k]:
        ls.append(df.loc[k, ['name','view_x','view_y']])
        
    else:
        if (df.loc[:, 'view_y'][k] == 'Ext') and (df.loc[:, 'aircraft_type'][k] in ls_model) and (df.loc[:, 'aircraft_type'][k] != df.loc[:, 'type'][k]):
            ls2.append(df.loc[k, ['name','aircraft_type','type']])

        else:
            if (df.loc[:, 'view_y'][k] == 'Int') and (df.loc[:, 'aircraft_manufacturer'][k] in ls_man) and (df.loc[:, 'manufacturer'][k] in ls_man) and (df.loc[:, 'aircraft_manufacturer'][k] != df.loc[:, 'manufacturer'][k]):
                ls3.append(df.loc[k, ['name', 'aircraft_manufacturer', 'manufacturer']])

            else: 
                if (df.loc[:, 'view_y'][k] == 'Int') and (df.loc[:, 'aircraft_type'][k] in ls_model) and (df.loc[:, 'type'][k] in ls_model) and (df.loc[:, 'aircraft_type'][k] != df.loc[:, 'type'][k]):
                    ls4.append(df.loc[k, ['name', 'aircraft_type', 'type']])
        

model_view = pd.DataFrame(ls)

metric_model_view = model_view.shape[0]/df.shape[0]

model_ext = pd.DataFrame(ls2)

metric_model_ext = model_ext.shape[0]/df[(df.loc[:,'view_y'] == 'Ext') & (df.loc[:, 'aircraft_type'].isin(ls_model))].shape[0]

model_int_man = pd.DataFrame(ls3)

metric_int_man = model_int_man.shape[0]/df[(df.loc[:, 'view_y'] == 'Int') & (df.loc[:, 'aircraft_manufacturer'].isin(ls_man)) & (df.loc[:, 'manufacturer'].isin(ls_man))].shape[0]

model_int_type = pd.DataFrame(ls4)

metric_int_type = model_int_type.shape[0]/df[(df.loc[:, 'view_y'] == 'Int') & (df.loc[:, 'aircraft_type'].isin(ls_model)) & (df.loc[:, 'type'].isin(ls_model))].shape[0]



```
